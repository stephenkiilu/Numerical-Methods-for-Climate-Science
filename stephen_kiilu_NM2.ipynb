```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=0
b=1
beta=2
alpha=0
N=200
h=(b-a)/N
x = np.linspace(a,b,N+1)

def f(x):
    return x*(1+x)
x = np.linspace(a,b,N+1)
exact= f(x)


# plt.figure(figsize=(9,8))
# plt.plot(x,exact)
# plt.title('Plot of exact' )
# plt.xlabel('x')
# plt.show()
```

```python
def k(x):
    return x**2
def f(x):
    return -2*x-6*x**2
```

```python
def steve(f,k):
    
    A=np.zeros((N+1,N+1))
    A[0,0]=1
    A[N,N]=1
    F=np.zeros(N+1)
    F[0]=alpha
    F[N]=beta
    for i in range(1,N):
        F[i] = h**2 * f(x[i])
        A[i,i]= k(x[i]-(h/2)) +k(x[i]+(h/2))
        A[i,i+1] = -k(x[i]+(h/2))
        A[i,i-1] =-k(x[i]-(h/2))
    return A,F
```

```python
A,F=steve(f,k)
F.shape
```

```python
U=np.linalg.solve(A,F)
```

```python
plt.figure(figsize=(9,8))
plt.plot(x,exact)
plt.plot(x[::5],U[::5],"*")
plt.title('Graph of exact and approximate solutions \n' )
plt.xlabel('x')
plt.ylabel('Heat conduction')
plt.legend(['Exact','Approximate'])
```
