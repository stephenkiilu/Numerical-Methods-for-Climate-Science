```python
mu=0.4
dx=0.05
bb=1
dt=(mu*dx**2)/bb
A=-10
B=10
a=0
b=1
N=int((B-A)/dx)
M=int((b-a)/dt)
```

```python
x=np.linspace(A,B,N+1)
t=np.linspace(a,b,M+1)

U0=np.zeros(N+1)
for i in range(N+1):
    U0[i]=max(1-x[i]**2,0)
```

```python
U=np.zeros((N+1,M+1))
U[0]=0
U[N]=0
U[:,0]=U0
```

```python
for n in range(1,M+1):
    for j in range(0,N):
        U[j,n]=(1-2*mu)*U[j,n-1]+mu*(U[j-1,n-1]+U[j+1,n-1])
```
