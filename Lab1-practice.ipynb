```python
def fun2(t,y):
    return (2-y**2)/(1+t**2)
   

def  EulerMethod(a,b,n,h,fun2):
    h = (b-a)/n
    t = np.linspace(a,b,n+1)
    y = np.zeros([n+1])
    y[0] = 1
    for i in range(1,n+1):
        y[i] = y[i-1] + h*fun2(t[i-1],y[i-1])
    
    return t,y
```

```python
def fun3(t,y):
    return (1-alpha)*s0/(4*c)-(e*sigma*y**4)/c

```

## Euler method

```python
def  EulerMethod(a,b,h,N):
    t = np.linspace(a,b,N+1)
    Y = np.zeros([N+1])
    Y[0] = 300
    for i in range(1,N+1):
        Y[i] = Y[i-1] + h*fun3(t[i-1],Y[i-1])
    
    return t,Y


```

```python
alpha=0.3
s0=1367
e=0.6
c=85
sigma=5.67*10**-8
a=0
b=500
h = 14
N=int((b-a)/h)
t,euler=EulerMethod(a,b,h,N)
```

```python
plt.plot(t,euler)
```

## Mid point method

```python
def  midpoint(a,b,fun3):
    t = np.linspace(a,b,N+1)
    tt=np.zeros([N+1])
    y2 = np.zeros([N+1])
    y_2= np.zeros([N+1])
    y2[0] = 300
    for i in range(0,N):
        tt[i]=t[i]+0.5*h
        y_2[i]=y2[i]+(0.5*h)*fun3(t[i],y2[i])
        y2[i+1] = y2[i] + h*fun3(tt[i],y_2[i])
    
    return t,y2
```

```python
a=0
b=500
h = 14
N=int((b-a)/h)
t,mid=midpoint(a,b,fun3)
```

```python
plt.plot(t,mid)
```

### Range Kutta 4

```python
def RK4(t,y,fun3):
    t = np.linspace(a,b,N+1)
    y3 = np.zeros([N+1])
    y3[0] = 300
    for i in range(0,N):
        k1 = fun3(t[i],y3[i])
        k2 = fun3(t[i]+0.5*H,y3[i]+0.5*H*k1)
        k3 = fun3(t[i]+0.5*H,y3[i]+0.5*H*k2)
        k4 = fun3(t[i]+H,y3[i]+H*k3)
        y3[i+1] = y3[i] + H/6*(k1+ 2*k2 + 2*k3 + k4)
    return t,y3
    
    
        
    

    
```

```python
a=0
b=500
H = 14
N=int((b-a)/h)
t,Rk4=RK4(t,y,fun3)
```

```python
plt.plot(t,Rk4)
```

## Comparing RK4, Euler Method and Mid point

```python
t,Rk4=RK4(t,y,fun3)
t,mid=midpoint(a,b,fun3)
t,euler=EulerMethod(a,b,h,N)
plt.plot(t,euler)
plt.plot(t,mid)
plt.plot(t,Rk4)
plt.legend(['Euler',"Mid-point","RK4"])
```
