```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=14
a=0
b=310
n=int((b-a)/h)
t=np.linspace(0,310,n+1)
S=1367
alpha=0.3
C=85
sigma=5.67*10**-8
e=0.6
T_eq=(((1-alpha)*S)/(4*e*sigma))**0.25
T_b=10*np.exp(((-4*e*sigma*T_eq**3)/(C))*t)



Y_exact=T_b+T_eq
#plt.plot(t,Y_exact)
```

```python
def fun(t,y):
        return (1-alpha)*S/(4*C)-(e*sigma*y**4)/C
    
```

```python
def RK(t):
    tt=np.zeros(n+1)
    y1=np.zeros(n+1)
    y=np.zeros(n+1)
    y[0]=299.57966642
    for i in range(0,n):
        K1=fun(t[i],y[i])
        tt[i]=t[i]+0.75*h
        y1[i]=y[i]+0.75*h*K1
        K2=fun(tt[i],y1[i])
        y[i+1]=y[i]+(1/3)*h*(K1+2*K2)
    return t,y
```

```python

a=0
b=310
h=14
n=int((b-a)/h)
t=np.linspace(0,310,n+1)
t,y_app= RK(t)
```

```python
plt.figure(figsize=(9,8))
plt.plot(t,Y_exact)
plt.plot(t,y_app)
plt.title('Plot of exact and approximate solutions \n' )
plt.xlabel('time')
plt.ylabel('Temperature')
plt.legend(['Exact','Approximate'])
plt.show()
```
